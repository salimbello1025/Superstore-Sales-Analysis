```python
# ================================================
# SALES DATASET EXPLORATION II — SUPERSTORE
# Student: Salim Bello Muhammad
# Program: 3MTT Data Analytics — NextGen Cohort
# Date: June 2026
# ================================================

import pandas as pd

# Load dataset
path = r'C:\Users\User\Downloads\Superstore (2).xlsx'

orders = pd.read_excel(path, sheet_name='Orders')
customers = pd.read_excel(path, sheet_name='Customer Name')
returns = pd.read_excel(path, sheet_name='Returns')
users = pd.read_excel(path, sheet_name='Users')

# Clean column names
orders.columns = orders.columns.str.replace(' ', '_')
customers.columns = customers.columns.str.replace(' ', '_')

print("✅ Data loaded successfully!")
print(f"Orders: {orders.shape}")
print(f"Customers: {customers.shape}")
print(f"Returns: {returns.shape}")
print(f"Users: {users.shape}")
```

```python
# Question 4: What is the total sales revenue?
print("Q4 - Total Sales Revenue")
print("=" * 50)
total_sales = orders['Sales'].sum()
print(f"Total Sales Revenue: ${total_sales:,.2f}")
```

```python
# Question 5: What is the total sales by Region?
print("Q5 - Total Sales by Region")
print("=" * 50)
q5 = orders.groupby('Region')['Sales'].sum().reset_index()
q5.columns = ['Region', 'Total_Sales']
q5 = q5.sort_values('Total_Sales', ascending=False)
q5['Total_Sales'] = q5['Total_Sales'].round(2)
q5
```

```python
# Question 6: What is the total profit by Product Category?
print("Q6 - Total Profit by Product Category")
print("=" * 50)
q6 = orders.groupby('Product_Category')['Profit'].sum().reset_index()
q6.columns = ['Product_Category', 'Total_Profit']
q6 = q6.sort_values('Total_Profit', ascending=False)
q6['Total_Profit'] = q6['Total_Profit'].round(2)
q6
```

```python
# Question 7: What is the average discount by Customer Segment?
print("Q7 - Average Discount by Customer Segment")
print("=" * 50)
q7 = orders.groupby('Customer_Segment')['Discount'].mean().reset_index()
q7.columns = ['Customer_Segment', 'Average_Discount']
q7 = q7.sort_values('Average_Discount', ascending=False)
q7['Average_Discount'] = q7['Average_Discount'].round(2)
q7
```

```python
# Question 8: Who are the top 5 most profitable products?
print("Q8 - Top 5 Most Profitable Products")
print("=" * 50)
q8 = orders.groupby('Product_Name')['Profit'].sum().reset_index()
q8.columns = ['Product_Name', 'Total_Profit']
q8 = q8.sort_values('Total_Profit', ascending=False).head(5)
q8['Total_Profit'] = q8['Total_Profit'].round(2)
q8
```

```python
# Question 9: Which regions have total sales above 500,000?
print("Q9 - Regions With Total Sales Above 500,000")
print("=" * 50)
q9 = orders.groupby('Region')['Sales'].sum().reset_index()
q9.columns = ['Region', 'Total_Sales']
q9 = q9[q9['Total_Sales'] > 500000]
q9 = q9.sort_values('Total_Sales', ascending=False)
q9['Total_Sales'] = q9['Total_Sales'].round(2)
q9
```

```python
# Question 10: Total orders and sales per Ship Mode
print("Q10 - Total Orders and Sales per Ship Mode")
print("=" * 50)
q10 = orders.groupby('Ship_Mode').agg(
    Total_Orders=('Ship_Mode', 'count'),
    Total_Sales=('Sales', 'sum')
).reset_index()
q10 = q10.sort_values('Total_Sales', ascending=False)
q10['Total_Sales'] = q10['Total_Sales'].round(2)
q10
```
